Before everything, we should know that:

If the regression equation is linear, it is of the form:
\begin{equation*}
μY|x = α + βx
\end{equation*}

For multiple linear regression, it is of the form:
\begin{equation*}
μY|x_1, x_2, ..., x_k = β_0 + β_1x_1 + β_2x_2 + ··· + β_kx_k
\end{equation*}

The algorithm we are going to do, is for multiple linear regression.

```
import numpy  #NumPy is used commonly in Python for maths
import matplotlib  #matplotlib is used for data visualization
import matplotlib.pyplot

X = 2 * numpy.random.rand(50, 1)
y = 5 + 10 * X + 3 * numpy.random.randn(50, 1)

matplotlib.pyplot.xlabel("x")
matplotlib.pyplot.ylabel("y", rotation=0)
matplotlib.pyplot.plot(X, y, "b.")
matplotlib.pyplot.axis([0, 2, 0, 40])
matplotlib.pyplot.show()  #We created some random data, let's see how they looked.
```

We can see that "mainly" we are generating data "around" y = 10x + 10. We were adding random "noises" on it.

The least squares estimates of the multiple regression coefficients are:
\begin{equation*}
B = (X'X)^{−1} X'Y
\end{equation*}
Below is the code.

```
X_b = numpy.c_[numpy.ones((50, 1)), X]  # adding x0 = 1 to each instance
B = numpy.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
print(B)
```

Now we have our model!

We make a prediction then, to show how good it works.

```
X_test = numpy.array([[0], [1], [2]]) #2 points would be enough
X_test_b = numpy.c_[numpy.ones((3, 1)), X_test]  # adding x0 = 1 to each instance
y_test = X_test_b.dot(B)
print(y_test)
```

Now plot the test predictions of our model:

```
matplotlib.pyplot.xlabel("x")
matplotlib.pyplot.ylabel("y", rotation=0)
matplotlib.pyplot.plot(X_test, y_test, "r.")
matplotlib.pyplot.plot(X_test, y_test, "y-")
matplotlib.pyplot.plot(X, y, "b.")  #accompany with our data
matplotlib.pyplot.axis([0, 2, 0, 40])
matplotlib.pyplot.show()
```
